```python
import pandas as pd
import json
```

```python
df = pd.read_json('covid19data.json')
```

```python
df
```

```python
df_prep = df[df['code'].isin(['DEP-75','DEP-77','DEP-78','DEP-91','DEP-92','DEP-93','DEP-94','DEP-95'])]
```

```python
df_prep = df_prep[['date','nom','deces','hospitalises','reanimation']]
```

```python
df_prep
```

```python
df_prep.quantile([.3,.6,.8], axis = 0)
```

```python
quantile_dis = [[3,1,1,1,2,1,2,1],[3,1,2,1,2,3,1,1],[3,1,1,1,2,3,1,1]]
```

```python
df_prep['deces_q'],df_prep['hospitalises_q'],df_prep['reanimation_q']=quantile_dis[0],quantile_dis[1],quantile_dis[2]
```

```python
df_prep['hospitalises_q']=[3,1,2,1,2,3,1,1]
```

```python
df_prep['reanimation_q']=[3,1,1,1,2,3,1,1]
```

```python
df_prep_q=df_prep
```

```python
df_prep_q['deces_q']=df_prep_q['deces_q']*0.5
df_prep_q['hospitalises_q']=df_prep_q['hospitalises_q']*0.2
df_prep_q['reanimation_q']=df_prep_q['reanimation_q']*0.3
```

```python
df_prep_q['sum_q']=df_prep_q['deces_q']+df_prep_q['hospitalises_q']+df_prep_q['reanimation_q']
```

```python
df_prep_q
```

```python
df_prep_q["covid_state"] = pd.np.nan
```

```python
mask = (df_prep_q['sum_q'] <= 3)
df_prep_q['covid_state'][mask] = "high"
mask = (df_prep_q['sum_q']>1) & (df_prep_q['sum_q'] <= 2)
df_prep_q['covid_state'][mask] = "medium"
mask = (df_prep_q['sum_q']>0) & (df_prep_q['sum_q'] <= 1)
df_prep_q['covid_state'][mask] = "low"
```

```python
df_prep_q
```

```python
## create new columns
df_prep_q["place_name"] = pd.np.nan
df_prep_q["place_address"] = pd.np.nan
df_prep_q["place_populartimes"] = pd.np.nan
```

```python
# list of cities
cities = ['Paris','Hauts-de-Seine','Seine-Saint-Denis']
```

```python
for city in cities:
    for index, row in df_prep_q.iterrows():
        if row['nom']==city:
            with open('data/'+city+'.json') as json_file:
                data = json.load(json_file)
                for d in data:
                    row['place_name']=d['name']
                    row['place_address']=d['address']
                    row['place_populartimes']=d['populartimes'][1]['data']
                    a_series = pd.Series(row, index = df_prep_q.columns)
                    df_prep_q = df_prep_q.append(a_series, ignore_index=True)
```

```python
df_prep_q_places=df_prep_q.dropna()
```

```python
df_prep_q_places
```

```python
# create place_populartimes hours
df_prep_q_places_h=df_prep_q_places
for _ in range(0,24):
    df_prep_q_places_h['place_populartimes_'+str(_)]=pd.np.nan
```

```python
columns_place_populartimes_h = [x for x in list(df_prep_q_places_h.columns) if "place_populartimes_" in x]
```

```python
for index, row in df_prep_q_places_h.iterrows():
    for r,z in zip(list(row['place_populartimes']),columns_place_populartimes_h):
        df_prep_q_places_h.at[index,z]=r
```

```python
df_prep_q_places_h
```

```python
# add nearby hospital and event random

df_prep_q_places_h_nb=df_prep_q_places_h
df_prep_q_places_h_nb['hospital-nearby']=pd.np.random.randint(0,2, size=len(df_prep_q_places_h_nb))
df_prep_q_places_h_nb['event-nearby']=pd.np.random.randint(0,2, size=len(df_prep_q_places_h_nb))
```

```python
df_prep_q_places_h_nb
```
